```python
from bs4 import BeautifulSoup as bs
import requests
from tqdm import  tqdm_notebook, _tqdm_notebook
import pandas as pd
import numpy as np
```

# 匯入各區座標

```python
data = pd.read_csv('area_location_un.csv')
data = data.drop(columns = ['主要縣市', 'Unnamed: 4','Unnamed: 5','Unnamed: 6'])
data
```

# function 區

```python
def filters(word):
    word = word.replace('N','')
    word = word.replace('E','')
    word = word.replace('’','')
    word = word.replace('”','')
    return int(word)
```

# 把座標變成可計算座標

```python
for x in tqdm_notebook(range(0, len(data))):
    data.loc[x, 'Longitude (經度)'] = filters(data.loc[x, 'Longitude (經度)'])
    data.loc[x, 'Latitude(緯度)'] = filters(data.loc[x, 'Latitude(緯度)'])
    
data    
```

# 處理成所需格式

```python
data['縣市'] = 0
data['地區'] = 0
for x in tqdm_notebook(range(0, len(data))):
    data.loc[x, '縣市'] = (data.loc[x, '縣市各區'][0:3])
    data.loc[x, '地區'] = (data.loc[x, '縣市各區'][3:])
    
data

```

```python
data.drop(columns = ['縣市各區'])
data = data[['縣市', '地區', 'Longitude (經度)', 'Latitude(緯度)']]
data
```

```python
data.to_csv('area_location.csv')
```
